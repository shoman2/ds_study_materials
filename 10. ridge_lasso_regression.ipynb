```python
import pandas as pd
import numpy as np
```

# Read and preprocess the data

- Read the data

```python
# Read the data
data = pd.read_csv('data/bikeshare.csv')
```

```python
# Explore the first 10 rows
data.head(10)
```

- Handle the datetime variable.

```python
# Add 'year', 'month' and 'hour'
datetime = pd.DatetimeIndex(data['datetime'])
data['year'] = datetime.year
data['month'] = datetime.month
data['hour'] = datetime.hour

# Drop 'datetime'
data.drop('datetime', axis=1, inplace=True)
```

```python
# Explore the first 10 rows
data.head(10)
```

- Change the variable name.

```python
# Rename the variable 'count'
# 'count' is a method, so it's best to name that column something else 
data.rename(columns={'count':'total'}, inplace=True)
```

```python
# Explore the first 10 rows
data.head(10)
```

- Drop some variables

```python
# Drop the variables named 'casual' and 'registered'
# 'causal' and 'registered' cannot be used to predict the target variable 'total'
# (You have to understand the data generation background!)
data.drop(['casual', 'registered'], axis=1, inplace=True)
```

```python
# Explore the first 10 rows
data.head(10)
```

- Add the derivative variable.

```python
# Add derivative variable "daytime"
data['daytime'] = ((data.hour > 6) & (data.hour < 21)).astype(int)
```

```python
# Explore the first 10 rows
data.head(10)
```

- Convert nominal variables into several binary dummy variables.

```python
# Convert 'season'
season_dummies = pd.get_dummies(data.season, prefix='season')
season_dummies.drop(season_dummies.columns[0], axis=1, inplace=True)
data = pd.concat([data, season_dummies], axis=1)

# Drop 'season'
data.drop('season', axis=1, inplace=True)
```

```python
# Explore the first 10 rows
data.head(10)
```

```python
# Convert 'hour'
hour_dummies = pd.get_dummies(data.hour, prefix='hour')
hour_dummies.drop(hour_dummies.columns[0], axis=1, inplace=True)
data = pd.concat([data, hour_dummies], axis=1)

# Convert 'hour'
data.drop('hour', axis=1, inplace=True)
```

```python
# Explore the first 10 rows
data.head(10)
```

- Check variable names and data types

```python
print(data.columns)
```

```python
print(data.dtypes)
```

# Data split

```python
from sklearn.model_selection import train_test_split
```

```python
# All variables except 'total' (target variable) => X
# 'total' (target variable) => Y
X = data.drop('total', axis=1)
Y = data['total']
```

```python
# Split the data: train set 70%, test_set 30%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

# Simple linear regression vs. Ridge vs. Lasso

```python
# Import classes
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

- Simple linear regression

```python
# Fit the linear regression model
linreg = LinearRegression()
linreg.fit(X_train, Y_train)

# Calculate Y_train_hat and Y_test_hat
Y_train_hat = linreg.predict(X_train)
Y_test_hat  = linreg.predict(X_test)

# Evaluate the model
rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_hat))
rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_hat))
r2_train = linreg.score(X_train, Y_train)
r2_test = linreg.score(X_test, Y_test)
print("Train RMSE: ", rmse_train)
print("Test RMSE: ", rmse_test)
print("Train R^2: ", r2_train)
print("Test R^2: ", r2_test)
```

- Ridge

```python
# Fit the Ridge regression model
ridge = Ridge(alpha=0.5, normalize=False)
ridge.fit(X_train, Y_train)

# Calculate Y_train_hat and Y_test_hat
Y_train_hat = ridge.predict(X_train)
Y_test_hat  = ridge.predict(X_test)

# Evaluate the model
rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_hat))
rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_hat))
r2_train = ridge.score(X_train, Y_train)
r2_test = ridge.score(X_test, Y_test)
print("Train RMSE: ", rmse_train)
print("Test RMSE: ", rmse_test)
print("Train R^2: ", r2_train)
print("Test R^2: ", r2_test)
```

- Lasso

```python
# Fit the Lasso regression model
lasso = Lasso(alpha=1.0, normalize=False)
lasso.fit(X_train, Y_train)

# Calculate Y_train_hat and Y_test_hat
Y_train_hat = lasso.predict(X_train)
Y_test_hat  = lasso.predict(X_test)

# Evaluate the model
rmse_train = np.sqrt(mean_squared_error(Y_train, Y_train_hat))
rmse_test = np.sqrt(mean_squared_error(Y_test, Y_test_hat))
r2_train = lasso.score(X_train, Y_train)
r2_test = lasso.score(X_test, Y_test)
print("Train RMSE: ", rmse_train)
print("Test RMSE: ", rmse_test)
print("Train R^2: ", r2_train)
print("Test R^2: ", r2_test)
```
